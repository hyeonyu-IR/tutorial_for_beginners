# Simple segmentation (thresholding)

```python
import numpy as np
try:
 from skimage.filters import threshold_otsu
except Exception:
 threshold_otsu=None
np.random.seed(42)
img=np.random.normal(50,20,(128,128)).astype('float32')
th=threshold_otsu(img) if threshold_otsu else 60.0
mask=(img>=th).astype('uint8')
float(th), int(mask.sum())
```
