# NumPy Basics (Radiology)
HU windowing demo for a simulated CT slice.

```python
import numpy as np
np.random.seed(42)
slice_hu = np.random.normal(40, 100, (256,256))
center,width=50,150
low,high=center-width/2,center+width/2
windowed=np.clip(slice_hu, low, high)
slice_hu.mean(), slice_hu.min(), slice_hu.max()
```
